```python
attribute_pair = "she_he"
# df_00 = pd.read_csv('original/female_male_full_pairs.csv')
# df_11 = pd.read_csv('original/male_female_full_pairs.csv')

df_00 = pd.read_csv('female_male_full_pairs_'+attribute_pair+'.csv')
df_11 = pd.read_csv('male_female_full_pairs_'+attribute_pair+'.csv')

```

```python
df_00 = df_00.rename(columns={'original_old_word': 'female', 'new_word': 'male'})
df_11 = df_11.rename(columns={'new_word': 'female', 'original_old_word': 'male'})

df_00 = df_00[['female','male','logits']]
df_11 = df_11[['female','male','logits']]
```

```python
df_0 = df_00[~df_00.female.isin(df_11.male)]
df_0
df_1 = df_11[~df_11.male.isin(df_00.female)]
df_1
```
